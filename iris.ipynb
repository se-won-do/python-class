```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
x_input = pd.DataFrame(data['data'], columns=data['feature_names'])
x_input.head()
```

```python
x_input.info()
```

```python
y_label = pd.DataFrame(data['target']) #0: setsota 1:versicolor 2:virginica
y_label.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_input, 

                                                    y_label , 

                                                    test_size=0.4, 

                                                    shuffle=False, 

                                                    random_state=42)



print('X_train shape:', X_train.shape)

print('X_test shape:', X_test.shape)

print('y_train shape:', y_train.shape)

print('y_test shape:', y_test.shape)
```

##02.모델

```python
from sklearn.linear_model import LogisticRegression


# 모델 생성 및 선택
model=LogisticRegression()

# 학습
hist=model.fit(X_train, y_train.values.ravel())

# 평가

train_score=model.score(X_train, y_train) #정확도
print("train score=",train_score)
y_pred = model.predict(X_train)
#print(y_pred) #예측값 확인
test_score= model.score(X_test, y_test)
print("test score=",test_score)

```

## 2. 데이터 불러오기

```python
from pycaret.datasets import get_data
dataset = get_data('iris')
```

```python
dataset['species'].unique()
```

```python
from sklearn.model_selection import train_test_split #데이터를 나누기 위해
```

```python
train, test = train_test_split(dataset, test_size = 0.2, random_state = 0, shuffle = True)
```

```python
train.shape, test.shape
```

```python
train.columns
```

## 3. 환경 구축 [setup()](https://pycaret.readthedocs.io/en/latest/api/classification.html#module-pycaret.classification)

```python
# Colab 사용시 아래 코드 실행
from pycaret.utils import enable_colab
enable_colab()
```

```python
from pycaret.classification import *
clf = setup(data = train, target = 'species', session_id=0)#randomstate랑 같은 역할->session_id=0
```

- session_id : 결과 재생산을 위한 랜덤 시드

- Target Type : 타겟 변수의 형태 (Binary or Multiclass)

- Transformed Train Set : clf 환경에서 변경된 훈련 데이터

- Transformed Test Set : clf 환경에서 변경된 시험 데이터

## 4. 모델 비교 [compare_models()](https://pycaret.readthedocs.io/en/latest/api/classification.html#pycaret.classification.compare_models)

- sort: 해당 지표 기준으로 정렬 (가장 좋은 모델이 최상단)
- n_select: 가장 좋은 n개의 모델을 골라서 반환
- fold: 교차 검증 시 사용할 폴드 개수
- budget_time: 명시된 시간 동안만 함수를 실행 (분 단위)
- include: 명시된 모델만 학습
- exclude: 명시된 모델 학습에서 제외
- turbo: True일 시, True인 모델만 학습
- models() 함수를 사용해 훈련 가능한 알고리즘 확인 가능

```python
models()#사용가능 모델확인 터보true하면 터보true만 학습하게 됨. 
```

```python
best = compare_models(n_select = 3,sort="acc")#상위 3개만 반환해서 best에 입력sort: str='Accuracy'디폴트여서 이거 기준 상위 3개 나열함. 
```

```python
best
```

## 5. 모델 튜닝 [tune_model()](https://pycaret.readthedocs.io/en/latest/api/classification.html#pycaret.classification.tune_model)

```python
tuned_model = [tune_model(i) for i in best]
```
